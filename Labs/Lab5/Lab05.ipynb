```python
data = pd.read_csv('https://raw.githubusercontent.com/patrickmlong/Breast-Cancer-Wisconsin-Diagnostic-DataSet/master/data.csv')
```

```python
print(f'shape {data.shape}\n')
print(f'column headers \n{data.columns}\n')
print(f'first 5 items: \n{data[:5]}\n')

print(f'mean of columns: \n{data.mean(axis=0)}\n')

rep_dict = {'M': 1, 'B': 0}
data.replace(to_replace = rep_dict, inplace=True)
print(f'updated first 5 items: \n{data[:5]}\n')
```

```python
headers = data.columns
new_headers = headers.drop(['id', 'diagnosis', 'Unnamed: 32'])
cancer_x = data[new_headers]
print(f'cancer_x first 5 items: \n{cancer_x[:5]}\n')

mins, maxes = cancer_x.min(axis=0), cancer_x.max(axis=0)
cancer_x_normalized = (cancer_x - mins) / (maxes-mins)

print(f'cancer_x normalized first 5 items: \n{cancer_x_normalized[:5]}\n')
```

## Task 2: Create and Visualize a Covariance Matrix

1. Use the [Dataframe](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html) documentation to find out how to compute the pairwise covariance of the columns, store the resulting covariance matrix, and print its shape.
2. Use the [matplotlib](https://matplotlib.org/3.1.1/gallery/images_contours_and_fields/image_annotated_heatmap.html) documentation to visualize the covariance matrix with a heatmap.

```python
cov = cancer_x_normalized.cov()
print(f'covariance matrix shape {cov.shape}')

print("Expected Covariance matrix shape: (30, 30)")
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
im = ax.imshow(cov, cmap='twilight_r')

headers = cov.columns

ax.set_xticks(np.arange(len(headers)))
ax.set_yticks(np.arange(len(headers)))
ax.set_xticklabels(headers)
ax.set_yticklabels(headers)

ax.set_title('Covariance matrix heatmap')

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

_ = fig.colorbar(im, ax=ax)
```

1. Use [numpy.linalg.eig](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.eig.html) to compute the eigenvalues and right eigenvectors of the covariance matrix.  Print the eigen_vectors shape and eigen_values.

```python
eigval, eigvec = np.linalg.eig(cov)
print(f'eigen values:\n{eigval}')
print(f'eigen vectors shape:\n{eigvec.shape}')

transformed_cancer_x = cancer_x @ eigvec[:, :2]

ax.set_title('PCA projected')
ax.set_xlabel('first principal component')
ax.set_ylabel('second principal component')

fig, ax = plt.subplots()
scatter = ax.scatter(transformed_cancer_x.iloc[:, 0], transformed_cancer_x.iloc[:, 1], 
                     c=data['diagnosis'], cmap='twilight_r')

handles, labels = scatter.legend_elements()

fig.legend(handles, ['Benign', 'Malignant'])
fig.colorbar(scatter, ax=ax)


print('\n\n')
print ("Expected Eigen values:\n[0.33133 0.10785 0.04439 0.04001 0.0255  0.01917 0.00986 0.00743 0.00617 0.0059  0.0053  0.00473 0.0041  0.00296 0.00168 0.00161 0.00115 0.00094 0.00086 0.00066 0.00061 0.00049 0.      0.00002 0.00003 0.00012 0.00021 0.00027 0.00032 0.00033]")
print ("Expected Eigen vectors shape:\n(30, 30)")
print('''Expected Transformed data:
           0         1
0 -2.590770  0.762974
1 -1.666057 -0.220868
2 -2.158371  0.226378
3 -2.204565  1.861168
4 -1.830577  0.033608''')
```
